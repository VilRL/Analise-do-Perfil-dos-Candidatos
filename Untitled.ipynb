# Analise de dados dos Candidatos do Estado de SP 2020

```python
import pandas as pd #importando a biblioteca pandas para analisar os dados
import matplotlib.pyplot as plt #importando a biblioteca matplotlib para fazer os gráficos
```

```python
vEleicoes = pd.read_csv("C:/Users/Kelliot/analise-dados/dados/consulta_cand_2020_SP.csv", sep=";", encoding='Latin-1') #carregando o arquivo de dados dos Candidatos do estado de SP
```

```python
ColunasSelecionada= ['NM_UE','NM_URNA_CANDIDATO',
       'SG_PARTIDO', 'DS_GENERO','DS_GRAU_INSTRUCAO'] #Aqui estamos selecionado apenas as colunas que nos interessa para usar
```

```python
vEleicoesSelecionado = vEleicoes.filter(items=ColunasSelecionada) #aqui criamos uma variavel pra armarzenar as colunas que vamos usar
```

```python
vEleicoesSelecionado.head() #Aqui podemos ver como ficou nossa tabela
```

```python
#Agora que temos uma noção dos dados que analisamos vamos montar eles em graficos
```

```python
plt.figure(figsize = (30,15))
vEleicoes["SG_PARTIDO"].hist(bins = 30, ec = "k", alpha = .5, color = "red")
plt.suptitle('Partido dos Candidatos do Estado de SP',  
             ha = 'center', 
             fontsize = 50, 
             weight = 'extra bold')

plt.xlabel("Partidos")
plt.ylabel("Quantidade de Candidatos")
```

```python
plt.figure(figsize = (25,15))
vEleicoes["DS_GRAU_INSTRUCAO"].hist(bins = 15, ec = "k", alpha = .6, color = "red")
plt.suptitle('Grau de Escolaridade dos Candidatos',  
             ha = 'center', 
             fontsize = 50, 
             weight = 'extra bold')

plt.xlabel("Grau de Escolaridade")
plt.ylabel("Quantidade de Escolaridade")
```

```python
plt.figure(figsize = (15,10))
vEleicoes["DS_GENERO"].hist(bins = 5, ec = "k", alpha = .6, color = "red")
plt.suptitle('Grau de Genero dos Candidatos',  
             ha = 'center', 
             fontsize = 50, 
             weight = 'extra bold')

plt.xlabel("Genero dos Candidados")
plt.ylabel("Quantidade de Candidados")
```
